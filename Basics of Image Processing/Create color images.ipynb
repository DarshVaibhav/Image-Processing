## Merging these three shades

```python
img = []

# Red

for i in range(256):             # from Black to Red
    t = []
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)

for i in range(256,0,-1):        # from Red to Black
    t = []
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)
    
# Green
    
for i in range(256):             # from Black to Green
    t = []
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

for i in range(256,0,-1):        # from Green to Black
    t = []
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

    
#Blue    


for i in range(256):             # from Black to Blue
    t = []
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)

for i in range(256,0,-1):        # from Blue to Black
    t = []
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)
    
    

img = np.array(img)

plt.imshow(img)
```

```python
cv.imwrite('Transition.jpg',img)
```
